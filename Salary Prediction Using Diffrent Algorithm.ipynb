Salary Prediction Using Polynomial-Regression

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Importing the Dataset
dataset = pd.read_csv("Salary.csv")
X = dataset.iloc[:,:-1].values
y = dataset.iloc[:,1:].values

#Spliting dataset to training set & testing Set
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=0)    
```

```python
#Fiting the Polynomial Regression to the dataset
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
#accuracy of model depends on degree of polynomial
polynomial_reg = PolynomialFeatures(degree=4)
X_poly = polynomial_reg.fit_transform(X)
poly_reg = LinearRegression()
poly_reg.fit(X_poly,y)
```

```python
#visualizing the result for PolynomialRegression
plt.scatter(X,y,color="blue")
plt.plot(X,poly_reg.predict(polynomial_reg.fit_transform(X)),color="red")
plt.title("Salary Prediction Model(PolynomialRegression)")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()
```

```python
#Predict
poly_pred = poly_reg.predict(polynomial_reg.fit_transform([[9.2]]))
print("Predicted Salary on the Basis of years of experience(Polynomial_reg) : ",poly_pred)
```

Salary Prediction Using Decision tree Regression

```python
# Fit The Decision Tree Regressor
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor()
regressor.fit(X,y)
```

```python
#Visualizing For Decision Trees
X_grid = np.arange(min(X),max(X),0.01)
X_grid = X_grid.reshape((len(X_grid),1))

plt.scatter(X, y, color="red")
plt.plot(X_grid, regressor.predict(X_grid), color="blue")
plt.title("Decision Tree Regressor")
plt.xlabel("Position")
plt.ylabel("Salary")
plt.show()
```

```python
#Make Prediction 
y_pred2 = regressor.predict([[9.2]])
print("Predicted Salary on the Basis of years of experience(Decision Tree Regressor) : ",y_pred2)
```

```python
#Processing & Feature Scaling the Support Vector Regression
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
sc_y = StandardScaler()
X = sc_X.fit_transform(X)
y = sc_y.fit_transform(y)
```

```python
plt.scatter(X,y)
```

```python
from sklearn.svm import SVR
svr_reg = SVR(kernel='rbf')
svr_reg.fit(X,y)
```
